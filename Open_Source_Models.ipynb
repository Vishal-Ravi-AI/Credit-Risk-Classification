```python
from transformers import (
    AutoTokenizer, AutoModelForCausalLM,
    TrainingArguments, BitsAndBytesConfig
)
from peft import LoraConfig, prepare_model_for_kbit_training
from trl import SFTTrainer
from datasets import load_dataset
from accelerate import infer_auto_device_map, init_empty_weights

# Load your dataset file
dataset = load_dataset("json", data_files="credit_alpaca.jsonl")

# Format each row into a single prompt+response string
def format_example(example):
    prompt = f"{example['instruction']} {example['input']}\nAnswer:"
    return {"text": prompt + " " + example['output']}

# Apply formatting
formatted_dataset = dataset["train"].map(format_example)

```

```python
model_id = "meta-llama/Llama-3.1-8B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id, token=True)
tokenizer.pad_token = tokenizer.eos_token  # Set pad token
```

```python
def tokenize(example):
    return tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=512
    )

tokenized_dataset = formatted_dataset.map(tokenize, batched=True)

```

```python

lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
```

```python
from transformers import BitsAndBytesConfig, AutoModelForCausalLM
from peft import prepare_model_for_kbit_training

bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,  # or use load_in_4bit=True for even less memory
    llm_int8_enable_fp32_cpu_offload=False  # MUST BE FALSE for A100
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map={"": 0},  # force full model on GPU:0 (your A100)
    trust_remote_code=True
)

model = prepare_model_for_kbit_training(model)


```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./lora-llama3-credit-risk",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    save_strategy="epoch",
    logging_steps=10,
    remove_unused_columns=False,
    report_to=[]  # 👈 disables wandb and any other integrations
)


```

```python
trainer = SFTTrainer(
    model=model,
    train_dataset=tokenized_dataset,
    peft_config=lora_config,
    args=training_args
)

trainer.train()

```

```python
trainer.model.save_pretrained("./lora-llama3-credit-risk")
tokenizer.save_pretrained("./lora-llama3-credit-risk")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
from transformers import pipeline

base_model_id = "meta-llama/Llama-3.1-8B-Instruct"
adapter_dir = "./lora-llama3-credit-risk"

# Load base LLaMA model
tokenizer = AutoTokenizer.from_pretrained(base_model_id, trust_remote_code=True)
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=True
)

# Load LoRA adapter on top of base model
model = PeftModel.from_pretrained(base_model, adapter_dir)

# Create pipeline
llama_pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
from transformers import TrainingArguments
from trl import SFTTrainer
from peft import LoraConfig
from datasets import load_dataset
from transformers import DataCollatorForLanguageModeling
import os
os.environ["WANDB_DISABLED"] = "true"


dataset = load_dataset("json", data_files={"train": "credit_alpaca.jsonl"})


# Step 2: Format dataset (merge prompt and completion)
def format_prompt(example):
    return {"text": f"{example['instruction']}\n{example['input']}\nAnswer: {example['output']}"}


formatted_dataset = dataset["train"].map(format_prompt)

# Step 3: Set padding token if missing
openchat_tokenizer.pad_token = openchat_tokenizer.eos_token

# Step 4: Tokenize dataset
def tokenize(example):
    return openchat_tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=512,
    )

tokenized_dataset = formatted_dataset.map(tokenize, batched=True)

# Step 5: LoRA config
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

# Step 6: Training arguments
training_args = TrainingArguments(
    output_dir="./openchat-lora-credit-risk",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_steps=10,
    save_strategy="epoch",
    report_to="none",  # ✅ Disable all loggers including wandb
)

# Step 7: Train using SFTTrainer
from trl import SFTTrainer

trainer = SFTTrainer(
    model=openchat_model,
    train_dataset=tokenized_dataset,
    peft_config=lora_config,
    args=training_args,
    data_collator=DataCollatorForLanguageModeling(tokenizer=openchat_tokenizer, mlm=False)
)






trainer.train()
```
